```python
X = df.values
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=6)
y_predicted = km.fit_predict(X)
y_predicted
```

```python
df['cluster']=y_predicted
df.head()
df.to_csv("cluster.csv")
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
df4 = df[df.cluster==3]
df5 = df[df.cluster==4]
df6 = df[df.cluster==5]
```
